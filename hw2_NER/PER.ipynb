```python
import os
import pandas as pd
import csv
os.chdir('zh-NER-TF_modify') 
import main 

import win32file
win32file._setmaxstdio(2048)
```

```python
# Load news data and convert it to dataframe
NER_folder = os.path.dirname(os.getcwd())

PER_data_folder = os.path.join(NER_folder, 'PER')
if not os.path.exists(PER_data_folder):
    os.mkdir(PER_data_folder)

df = pd.read_csv(os.path.join(NER_folder, 'crawl_data_csv', 'news_data.csv'))
```

```python
df.head()
```

```python
# Put the result to zh-NER-TF model

"""notice: write!!!!!!!!!!!!!!!!!!! """

output_file = os.path.join(PER_data_folder, 'news_data_with_PER.csv')
k = 1
while(1):
    if not os.path.exists(output_file):
        with open(output_file, 'w', newline='', encoding = 'utf8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(list(df.columns) + ['PER'])
        
        index = 0
        for text in df['content_cn'].tolist():
            NER_result = main.do_NER(mode = 'demo', demo_model = '1521112368', demo_text = text)
            if NER_result == None:
                index += 1
                continue
            PER = NER_result['PER']
            PER_string = ','.join(PER)

            df_newrow = df.iloc[[index]]
            df_newrow.at[index, 'PER'] = PER_string

            with open(os.path.join(PER_data_folder, 'news_data_with_PER.csv'), 'a', newline='', encoding = 'utf8') as csvfile:
                df_newrow.to_csv(csvfile, header=False)

            index += 1
            print('Finish adding PER: news', index)
            
        break
        
    else:
        k += 1
        output_file = os.path.join(PER_data_folder, 'news_data_with_PER_' + str(k) + '.csv') 
```
