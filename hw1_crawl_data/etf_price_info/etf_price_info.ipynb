```python
import os
import pandas as pd
from selenium import webdriver
from datetime import datetime


etf_info_folder = 'etf_info'
etf_csvs = os.listdir(etf_info_folder)
etf_classes = []
etf_name_lists = []

for etf_csv in etf_csvs:
    print(etf_csv)
    etf_csv_dir = os.path.join(etf_info_folder, etf_csv)

    # make ETF classes
    list_split = etf_csv.split(' ')
    etf_class = ''
    for i in list_split:
        if i == 'ETF':
            break
        etf_class += i
    etf_classes.append(etf_class)

    df = pd.read_csv(etf_csv_dir)
    etf_date = {"etf_name": df["Symbol"], "inception": df["Inception"]}
    etf_date_df = pd.DataFrame(etf_date)
    # print(len(etf_date_df))

    for i in range(len(etf_date_df)):
        if '/' in etf_date_df["inception"][i]:
            d = datetime.strptime(etf_date_df["inception"][i], "%d/%m/%Y")
            etf_date_df["inception"][i] = d.strftime("%Y-%m-%d")

    print(etf_date_df.head())

    # before 2015-12-31
    mask = etf_date_df["inception"] <= "2015-12-31"
    etf_name_list = list(etf_date_df[mask]["etf_name"])
    etf_name_lists.append(etf_name_list)

    print('\nThe list of the ETF (before 2015-12-31).\n', etf_name_list)
    print('\nBefore 2015-12-31, the number of ETF in %s is %2d.\n' % (etf_csv, len(etf_name_list)))
```

```python
# The place we will direct our WebDriver to
baseurl = 'https://www.etf.com/'
etf_urls = []
homepage_urls = []

# Creating the WebDriver object using the ChromeDriver
options = webdriver.ChromeOptions()
# options.add_argument('--headless')
driver = webdriver.Chrome(chrome_options=options, executable_path='./chromedriver')

# Get the etf overview on etf.com
for item in etf_name_lists:
    item_urls = []
    for name in item:
        etf_url = baseurl + name + '#overview'
        item_urls.append(etf_url)
    etf_urls.append(item_urls)
# print(etf_urls)

# Get the homepage url
for item_urls in etf_urls:
    item_homepage_urls = []
    for url in item_urls:
        try:
            driver.get(url)
            # link = driver.find_elements_by_css_selector(
            #     '#fundSummaryData > div > div.rowText.fundHomepage > span > div > a')
            link = driver.find_element_by_link_text('Fund Home Page')
            item_homepage_urls.append(link.get_attribute("href"))
            print('Successfully get the homepage from ', url)
        except:
            item_homepage_urls.append('no url')
            print('Cannot get the homepage from ', url)
    homepage_urls.append(item_homepage_urls)
```

```python
# Check how many different company issue these ETF
etf_companies = []
for item_companies in homepage_urls:
    for url in item_companies:
        if url != 'no url':
            company = url.split('/')
            if company[2] not in etf_companies:
                etf_companies.append(company[2])
print(etf_companies)
```

```python
csv_dl_folder = 'price_csv_dl'
if not os.path.exists(csv_dl_folder):
    os.mkdir(csv_dl_folder)
for i in range(len(etf_name_lists)):
    etf_cls = etf_classes[i]
    etf_cls_folder = os.path.join(csv_dl_folder, etf_cls)
    if not os.path.exists(etf_cls_folder):
        os.mkdir(etf_cls_folder)
```

```python
def to_df(folder):
    files = os.listdir(folder)
    df_closes = []
    df_empty = True

    for file in files:
        etf_name = file.split('.')[0]
        file = os.path.join(folder, file)

        df = pd.read_csv(file)

        df_close = df[['Date', 'Close']]
        df_close.rename(columns={'Close': etf_name}, inplace=True)
        
        if df_empty:
            df_closes = df_close
            df_empty = False
            continue
        
        df_closes = pd.merge(df_closes, df_close, on='Date', how='outer')
    return df_closes


# Financials Equity folder
price_folders = os.listdir(csv_dl_folder)
cse_folder = price_folders[0]
price_folder = os.path.join(csv_dl_folder, cse_folder)

etf_df = to_df(price_folder)
etf_df.head(15)
```
